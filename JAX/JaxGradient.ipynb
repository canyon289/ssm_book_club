# Jax Gradients

```python
import matplotlib.pyplot as plt
import numpy as np
import jax.numpy as jnp
import jax
```

## A parabola

```python
x = np.linspace(-20, 20, 1000)

def y(x):
    return x**2
```

```python
plt.plot(x, y(x));
```

## What if we want the gradient?

```python
grad_y = jax.grad(y)
```

```python
grad_y(0.0), grad_y(2.0)
```

## Plot Gradients

```python
plt.plot(x, y(x));

grad_x = -4.0
x + grad_x
plt.plot(x + grad_x, (grad_y(grad_x)*x + y(grad_x)))
```

## Gradients for complex things

```python
def y(x):
    return 20*jnp.sin(x) + x**2

grad_y = jax.grad(y)
```

```python
plt.plot(x, y(x));

grad_x = 5.0
plt.plot(x + grad_x, (grad_y(grad_x)*x + y(grad_x)))
```

## Gradients in 2D

```python
def z(x,y):
    return 20*x*(4*y)

grad_z = jax.grad(z)

grad_z(1.0, 4.0), 4*20*4
```

```python
grad_z_res_y = jax.grad(z, argnums=1)

grad_z_res_y(1.0, 4.0), 20*4
```

## Value and Grad

```python
x = np.linspace(-20, 20, 1000)

def y(x):
    return x**2

jax.value_and_grad(y)(5.0)
```
